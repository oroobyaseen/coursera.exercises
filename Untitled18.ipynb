```python
def data_class(data_bath, size, class_node):
   datagen = ImageDataGenerator(rescale =1./255,
                                rotation_range=20,
                                width_shift_range=0.1,
                                height_shift_range=0.1,
                                shear_range=0.5,
                                zoom_range=(0.9, 1.1),
                                horizontal_flip=False,
                                vertical_flip=False,
                                fill_mode='constant',
                                cval=0,
                                validation_split=0.25    )
   classes = datagen.flow_from_directory(data_bath,
                                         target_size = size,
                                         batch_size = 32,
                                         seed=42,
                                         shuffle=True,
                                         class_mode = class_mode)
   return classes
```

```python
size = (156, 156)
class_mode = 'categorical'
```

```python
train_classes = data_class(train_bath, size ,"categorical")
```

```python
test_classes = test_datagen.flow_from_directory('/content/drive/MyDrive/New folder (4)/val',
                                            target_size = (156, 156),
                                            batch_size = 32,
                                            seed = 42,
                                            class_mode = 'categorical')
```

```python
# Import the VGG16 library as shown below and add preprocessing layer to the front of VGG
# Here we will be using imagenet weights

vgg19 = VGG19(input_shape=(156, 156, 3), weights='imagenet', include_top=False)
```

```python
# don't train existing weights
for layer in vgg19.layers:
    layer.trainable = False
```

```python
# our layers - you can add more if you want
x = Flatten()(vgg19.output)
```

```python
prediction = Dense(3, activation='softmax')(x)

# create a model object
model = Model(inputs=vgg19.input, outputs=prediction)
```

```python

# view the structure of the model
model.summary()
```

```python
plot_model(model=model, show_shapes =True)
```

```python
early_stop = EarlyStopping (monitor ="val_loss", mode = "min", verbose = 1, patience = 5)
```

```python
# tell the model what cost and optimization method to use
model.compile(
  loss='categorical_crossentropy',
  optimizer='rmsprop',
  metrics=['accuracy']
)
```

```python
from keras import callbacks
# fit the model
# Run the cell. It will take some time to execute
r = model.fit(
  train_classes,
  validation_data=test_classes,
  epochs=20,
  steps_per_epoch=len(train_classes),
  validation_steps=len(test_classes)
  #callbacks= [early_stop],
  #sullfle = True
)
```

```python
# plot the loss
plt.plot(r.history['loss'], label='train loss')
plt.plot(r.history['val_loss'], label='val loss')
plt.legend()
plt.show()
plt.savefig('LossVal_loss')
plt.title('VGG19-Model')

# plot the accuracy
plt.plot(r.history['accuracy'], label='train acc')
plt.plot(r.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
plt.savefig('AccVal_acc')
plt.title('VGG19-Model')
plt.show()
```

```python
def show_images(generator,y_pred=None):
    """
    Input: An image generator,predicted labels (optional)
    Output: Displays a grid of 9 images with lables
    """
    
    # get image lables
    labels =dict(((0,'Glioma'), (1,'Meningioma'), (2,'Pituitary tumor')))
    
    # get a batch of images
    x,y = generator.next()
    
    # display a grid of 9 images
    plt.figure(figsize=(10, 10))
    if y_pred is None:
        for i in range(9):
            ax = plt.subplot(2, 2, i + 1)
            idx = randint(0, 3064)
            plt.imshow(x[idx])
            plt.axis("off")
            plt.title("Class:{}".format(labels[np.argmax(y[idx])]))
                                                     
    else:
        for i in range(9):
            ax = plt.subplot(2, 2, i + 1)
            plt.imshow(x[i])
            plt.axis("off")
            plt.title("Actual:{} \nPredicted:{}".format(labels[np.argmax(y[i])],labels[y_pred[i]]))
    
# Display Train Images
show_images(train_data_gen)
```
